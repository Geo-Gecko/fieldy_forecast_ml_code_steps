```python
import rasterio
from rasterio.merge import merge
from rasterio.plot import show
import glob
import os
%matplotlib inline
```

```python
path = "20181003_20181005_2"
dir_list = os.listdir(path)
  
print("Files and directories in '", path, "' :") 
  
# print the list
print(dir_list)
```

```python
tff_lst  = []
for file in dir_list:
    filepath = os.path.join(path,file,"band8.tif")
#     print(filepath
    tff_lst.append(filepath)
```

```python
print(len(tff_lst))
print(tff_lst)
```

```python
src_files_to_mosaic = []

# Iterate over raster files and add them to source -list in 'read mode'
for tff in tff_lst:
#     print (tff)
    src = rasterio.open(tff)
    src_files_to_mosaic.append(src)

src_files_to_mosaic
```
